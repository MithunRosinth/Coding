```python
import matplotlib.pyplot as plt
```

```python
fig,ax1=plt.subplots()
time=[0,1,2,3,4,5,6]
co_con=[250,265,272,260,300,320,389]
temp=[14.1,15.5,16.3,18.1,17.3,19.1,20.2]
ax1.set_ylabel("CO2")
ax1.plot(time,co_con)
ax2=ax1.twinx()
ax2.set_ylabel("Temp (degC)")
ax2.plot(time,temp,'-g')
```
