```python
time=[0,1,2,3,4,5,6]
co_con=[250,265,272,260,300,320,389]
plt.plot(time,co_con,linestyle='dashed')
plt.ylabel("CO2 Concentration")
plt.xlabel("Time")
plt.title("Global Warming")
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pp=PdfPages('multipage.pdf')
time=[0,1,2,3,4,5,6]
co_con=[250,265,272,260,300,320,389]
temp=[14.1,15.5,16.3,18.1,17.3,19.1,20.2]
plt.plot(time,co_con,linestyle='dashed')
plt.plot(time,temp,'-g')
plt.ylabel("CO2 Concentration/Temperature")
plt.xlabel("Time")
plt.title("Global Warming")
plt.savefig(pp,format='pdf')
pp.savefig()
pp.close()
```
